# Map

Load data and merge country id's to the gapminder data. Country id's are found a file from here: https://github.ubc.ca/MDS-2020-21/DSCI_532_viz-2_students/tree/master/data

Note that there are differences in names in the country id's file above and the gapminder file and so some of the gapminder data is not shown on the map currently. We'll have to do some name cleaning to get all the countries to match. 

```python
import altair as alt
import pandas as pd
import numpy as np
from vega_datasets import data

world_map = alt.topo_feature(data.world_110m.url, 'countries')

# load processed gapminder data
df = pd.read_csv("../../data/processed/gapminder_processed.csv")
# select only the most current year
df = df.loc[df['year'] == 2018]

# load Joel's country id's
country_ids = pd.read_csv("../../data/raw/country-ids.csv")
country_ids

# merge country id's to gapminder data
df_merged = pd.merge(df, country_ids, how='left',
                    left_on='country', right_on='name')
```

```python
variable_interest = 'life_expectancy'

(alt.Chart(world_map).mark_geoshape()
.transform_lookup(
    lookup='id',
    from_=alt.LookupData(df_merged, key='id', fields=['name', variable_interest]))
.encode(tooltip=['name:O', variable_interest + ':Q'], color=variable_interest + ':Q')
.project(type='equalEarth'))
```
