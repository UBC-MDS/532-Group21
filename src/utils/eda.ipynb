```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import altair as alt
```

```python
dat = pd.read_csv("../../data/world-data-gapminder.csv")
```

```python
dat_asia = dat.query('region == "Asia" & year > 1970')
dat_asia["education_gender_ratio"] = dat_asia["years_in_school_women"] / dat_asia["years_in_school_men"]
dat_asia
```

```python
alt.Chart(dat_asia, title="Female/Male education mean ratio in Asia").mark_line().encode(
    x = "year",
    y = "mean(education_gender_ratio)",
    color = "income_group")
```

```python
dat_na = dat.query('region == "Americas" & year > 1975')
dat_na["education_gender_ratio"] = dat_na["years_in_school_women"] / dat_na["years_in_school_men"]
dat_na
```

```python
alt.Chart(dat_na, title = "Female/Male education mean ratio in America").mark_line().encode(
    x = "year",
    y = "mean(education_gender_ratio)",
    color = "income_group")
```

```python
alt.Chart(dat_na, title = "Female/Male education mean ratio in to child mortality rate in America").mark_circle(opacity=0.5).encode(
    x = alt.X("education_gender_ratio", scale=alt.Scale(domain=(0.6, 1.2))),
    y = "mean(child_mortality)",
    color = "income_group")
```

```python
alt.Chart(dat_asia, title = "Female/Male education mean ratio in to child mortality rate in America").mark_circle(opacity=0.5).encode(
    x = alt.X("education_gender_ratio", scale=alt.Scale(domain=(0, 1.2))),
    y = "mean(child_mortality)",
    color = "income_group")
```
