```python
import altair as alt
import pandas as pd
import numpy as np
import seaborn as sns
alt.data_transformers.disable_max_rows()

```

```python
# load data
df = pd.read_csv("../../data/raw/world-data-gapminder_raw.csv")
```

```python
df["education_ratio"] = df.years_in_school_men / df.years_in_school_women
```

```python
df = df.loc[df['year'] >= 1950]
```

```python
sns.heatmap(df.isnull(), cmap='viridis', cbar=False);
```

```python
filtered_df = df[df[['education_ratio', 'co2_per_capita', 'region']].notnull().all(1)]
```

```python
sns.heatmap(filtered_df.isnull(), cmap='viridis', cbar=False);
```

```python
filtered_df.to_csv("../../data/processed/gapminder_processed.csv")
```
