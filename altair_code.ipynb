```python
import altair as alt
import pandas as pd
import numpy as np
alt.data_transformers.disable_max_rows()

# load data
df = pd.read_csv("data/world-data-gapminder_processed.csv")
# select only the most current year
df = df.loc[df['year'] == 2018]

# specify variable of interest
variable_interest = 'children_per_woman'

(alt.Chart(df).mark_bar()
.encode(
    y = alt.Y('country', sort='-x'),
    x=variable_interest,
    color='country')
.transform_window(
    rank='rank(variable_interest)',
    sort=[alt.SortField(variable_interest, order='descending')])
.transform_filter(
    (alt.datum.rank < 6))
)
```

```python
data = pd.read_csv("data/world-data-gapminder_processed.csv")
```

```python
#Filters 
region = "Asia" 
# ['Asia', 'Europe', 'Africa', 'Americas', 'Oceania', 'unspecified']


sub_region = "Southern Asia" 
#       ['Southern Asia', 'Southern Europe', 'Northern Africa',
#        'Sub-Saharan Africa', 'Latin America and the Caribbean',
#        'Western Asia', 'Australia and New Zealand', 'Western Europe',
#        'Eastern Europe', 'South-eastern Asia', 'Northern America',
#        'Eastern Asia', 'Northern Europe', 'Melanesia', 'Central Asia',
#        'Micronesia', 'Polynesia', 'unspecified']


income_group = "Lower middle" 
#['Low', 'Upper middle', 'Lower middle', 'High', 'unspecified']


#Drag bar
population_size_lower, population_size_upper = 0, 100000000000

date_range_lower, date_range_upper = 2000, 2010
```

```python
#extract data frame
target_data_frame = data.query("region == @region and sub_region == @sub_region and income_group == @income_group")
target_data_frame = target_data_frame[target_data_frame["year"] > date_range_lower]
target_data_frame = target_data_frame[target_data_frame["year"] < date_range_upper]
target_data_frame = target_data_frame[target_data_frame["population"] > population_size_lower]
target_data_frame = target_data_frame[target_data_frame["population"] < population_size_upper]
```

```python
statistic_of_interest = 'income'
target_countries = target_data_frame.groupby("country").mean().sort_values(by = [statistic_of_interest], ascending=False).index[0:5].tolist()
df_plot = target_data_frame[target_data_frame["country"].isin(target_countries)]
```

```python
plot_title = statistic_of_interest + " of " + income_group + " countries in " + sub_region
plot_title = plot_title.title()
alt.Chart(df_plot, title=plot_title).mark_line().encode(
    x = "year",
    y = statistic_of_interest,
    color = "country")
```
